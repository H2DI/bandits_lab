```python
import bandits_lab.bandit_definitions as bands
import bandits_lab.algorithms as algs
```

```python
bandit_problem = bands.SimpleContBandPoly()
T = 1002

xs = np.linspace(0, 1, 100)
for _ in range(T):
    bandit_problem.play_arm(0.5)

plt.plot(xs, bandit_problem.compute_cum_reward(xs) / T)
```

```python
K = 10
discrete_alg = algs.Exp3(K)

alg = algs.AbstractCAB(
    K=K,
    discrete_alg=discrete_alg,
    discretizer=algs.MidpointDiscretizer(K),
    label="Test1",
)


for t in range(T):
    arm = alg.choose_arm()
    reward = bandit_problem.play(arm)
    alg.update(arm, reward)
```
